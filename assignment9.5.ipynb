Q1. How can you create a Bokeh plot using Python code?

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
from bokeh.plotting import figure, show,output_file
x = [1, 2, 3, 4, 5]
y = [4, 23, 0, 19, 52]
output_file("Bokeh Plot")
p = figure(title = "scatter plot")
p.xaxis.axis_label = "x axis"
p.yaxis.axis_label = "y axis"
p.scatter(x,y)
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

```python
"""
A Bokeh server is a way to create interactive plots that can be updated in real-time. It allows for interactions like hovering, clicking, and zooming.
To use a Bokeh server, you need to install the Bokeh library and create a Python script containing the code for the interactive plot.
"""
```

```python
from bokeh.io import curdoc
from bokeh.plotting import figure
from bokeh.server.server import Server
from tornado.ioloop import IOLoop
def modify_doc(doc):

    p = figure()
    p.circle([1, 2, 3], [4, 5, 6])
    doc.add_root(p)

# Create the Bokeh server
server = Server({'/': modify_doc}, io_loop=IOLoop(), allow_websocket_origin=["localhost:8080"])

# Start the server
server.start()

# Run the application
server.io_loop.start()
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

```python
"""
To embed a Bokeh plot into a web page or dashboard using Flask or Django, you need to integrate Bokeh into your web framework.

Install the necessary libraries:

Flask: pip install flask
Django: pip install django
"""
```

```python
from bokeh.plotting import figure
from bokeh.embed import components

# Create a figure object
p = figure()

# Add glyphs to the figure
p.circle([1, 2, 3], [4, 5, 6])

# Get the script and div components
script, div = components(p)
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()
```

```python
from django.shortcuts import render

def index(request):
    return render(request, 'index.html', {'script': script, 'div': div})
```
