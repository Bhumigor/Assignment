Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.


```python
"""
Scatter Plot: A plot that displays individual data points as markers on a two-dimensional plane. It helps in examining the 
relationship between two variables and identifying any correlation.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatter Plot")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
"""
A bar plot, also known as a bar chart or bar graph, is a visualization that uses rectangular bars to represent categorical 
data. The length of each bar corresponds to the value or frequency of the category it represents. Bar plots are particularly 
useful for comparing and displaying data across different categories or groups.

A bar plot is commonly used for the following purposes:

1.Comparisons: Bar plots allow for easy visual comparison of values or frequencies between different categories. The height or
length of each bar provides a clear representation of the data, making it straightforward to identify and compare magnitudes or trends.

2.Categorical Data: Bar plots are suitable for visualizing categorical data, where each bar represents a distinct category. 
They can display various types of categorical information, such as counts, frequencies, percentages, or proportions.

3.Grouped Data: Bar plots can effectively display and compare data across multiple groups or subcategories.By grouping the 
bars based on different criteria, such as time periods, demographic factors, or experimental conditions, it becomes possible to 
observe patterns, differences, or similarities across the groups.

4.Distribution: In some cases, a bar plot can be used to represent the distribution of a continuous variable by dividing it 
into discrete bins or intervals. The bars then indicate the frequency or count of data falling into each bin, providing insights 
into the shape or characteristics of the distribution.

5.Presentation and Communication: Bar plots are commonly used in presentations, reports, and publications due to their 
simplicity and effectiveness in conveying information. They provide a visually appealing way to present data and make it more 
accessible to a wider audience.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Profit of Company")
plt.title("Bar Plot")
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.ylabel("Company Name")
plt.xlabel("Profit of Company")
plt.title("Horizontal Bar Plot")
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
"""
A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of a dataset's 
distribution. It displays key statistical measures such as the median, quartiles, and potential outliers. The box plot is 
useful for understanding the range, spread, and skewness of the data.

Box plots are used for several purposes:

1.Visualizing Distribution: Box plots provide a concise visual summary of the distribution of a dataset. They offer insights 
into the central tendency, spread, and skewness of the data, including the presence of outliers.

2.Comparing Distributions: Box plots are useful for comparing the distributions of multiple datasets side by side. They enable 
quick comparisons of medians, quartiles, and ranges, facilitating the identification of differences and similarities between 
groups or categories.

3.Detecting Outliers: Box plots help in identifying potential outliers in a dataset. Outliers can be visually identified as 
data points beyond the whiskers, which may indicate unusual or extreme values.

4.Assessing Skewness: The asymmetry of a dataset can be visually assessed using box plots. If the median is not in the center 
of the box, it suggests a skew in the data towards the side where the box is shorter.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')

plt.show()
```
